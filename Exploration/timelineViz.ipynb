```python
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pgf import FigureCanvasPgf
import seaborn as sns

matplotlib.backend_bases.register_backend('pdf', FigureCanvasPgf)
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})
```

```python
with open('../data/timeline_18_12_23/timeline_full_raw.pkl', 'rb') as f:
    df = pickle.load(f)
    print(df.shape)
```

```python
# number of matchids in dataset
matchIds = df.index.get_level_values('matchId').unique().tolist()
print(len(matchIds))
```

```python
df.head(10)
```

```python
example_matchId = random.choice(matchIds)
```

```python
p1_health = df.loc[example_matchId]['participant1_health']
p1_healtMax = df.loc[example_matchId]['participant1_healthMax']
p1_movespeed = df.loc[example_matchId]['participant1_movementSpeed']
p1_currentGold = df.loc[example_matchId]['participant1_currentGold']
p1_physicalDamage = df.loc[example_matchId]['participant1_physicalDamageDone']
p1_totalGold = df.loc[example_matchId]['participant1_totalGold']
p1_totalDamageToChampions = df.loc[example_matchId]['participant1_totalDamageDoneToChampions']
p1_armor = df.loc[example_matchId]['participant1_armor']
timestamps = df.loc[example_matchId]['timestamp']
timestamps = timestamps / 60000
```

```python
ax = sns.lineplot(x=timestamps, y=p1_armor.values)
ax.set(xlabel='Time (minutes)', ylabel='Armor')
ax.set_title('Player 1 Armor')
plt.savefig('../thesis/images/participant1_armor.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_health.values)
ax.set(xlabel='Time (minutes)', ylabel='Health')
ax.set_title('Player 1 Health')
plt.savefig('../thesis/images/participant1_health.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_healtMax.values)
ax.set(xlabel='Time (minutes)', ylabel='Max Health')
ax.set_title('Player 1 Max Health')
plt.savefig('../thesis/images/participant1_maxHealth.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_movespeed.values)
ax.set(xlabel='Time (minutes)', ylabel='Movement Speed')
ax.set_title('Player 1 Movement Speed')
plt.savefig('../thesis/images/participant1_movementSpeed.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_currentGold.values)
ax.set(xlabel='Time (minutes)', ylabel='Current Gold')
ax.set_title('Player 1 Current Gold')
plt.savefig('../thesis/images/participant1_currentGold.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_physicalDamage.values)
ax.set(xlabel='Time (minutes)', ylabel='Physical Damage Done')
ax.set_title('Player 1 Physical Damage Done')
plt.savefig('../thesis/images/participant1_physicalDamage.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_totalGold.values)
ax.set(xlabel='Time (minutes)', ylabel='Total Gold')
ax.set_title('Player 1 Total Gold')
plt.savefig('../thesis/images/participant1_totalGold.pgf')
```

```python
ax = sns.lineplot(x=timestamps, y=p1_totalDamageToChampions.values)
ax.set(xlabel='Time (minutes)', ylabel='Total Damage Done To Champions')
ax.set_title('Player 1 Total Damage Done To Champions')
plt.savefig('../thesis/images/participant1_totalDamageToChampions.pgf')
```

```python
participants_totalGold = ['participant1_totalGold', 'participant2_totalGold', 'participant3_totalGold',
                          'participant4_totalGold',
                          'participant5_totalGold', 'participant6_totalGold', 'participant7_totalGold',
                          'participant8_totalGold',
                          'participant9_totalGold', 'participant10_totalGold']
participants_Ids = ['participant1_participantId', 'participant2_participantId', 'participant3_participantId',
                    'participant4_participantId',
                    'participant5_participantId', 'participant6_participantId', 'participant7_participantId',
                    'participant8_participantId',
                    'participant9_participantId', 'participant10_participantId']
df_totalGold = df.loc[example_matchId][participants_totalGold]
df_participantIds = df.loc[example_matchId][participants_Ids]
combined_df = pd.concat([timestamps, df_totalGold, df_participantIds], axis=1)
df_totalGold
```

```python
df_gold = pd.concat([timestamps, df_totalGold], axis=1)
df_gold = pd.melt(df_gold, id_vars=['timestamp'], value_vars=participants_totalGold)
df_gold['variable'] = df_gold['variable'].str.replace('_',
                                                      '')  # latex seems to have problems with underscores in labels
df_gold
```

```python
sns.color_palette("blend:royalblue,navy", 5)
```

```python
sns.color_palette("blend:lightcoral,firebrick", 5)
```

```python
blue_palette = sns.color_palette("blend:cornflowerblue,navy", 5)  # Adjust the number if needed
red_palette = sns.color_palette("blend:lightcoral,firebrick", 5)  # Adjust the number if needed

# Combine the palettes
combined_palette = blue_palette + red_palette
ax = sns.lineplot(data=df_gold, x='timestamp', y='value', hue='variable', palette=combined_palette)
ax.set(xlabel='Time (minutes)', ylabel='Gold')
ax.set_title('Total Gold of each player per minute')
handles, labels = ax.get_legend_handles_labels()
custom_labels = ['Player 1', 'Player 2', 'Player 3', 'Player 4', 'Player 5', 'Player 6', 'Player 7', 'Player 8',
                 'Player 9', 'Player 10']
ax.legend(handles=handles, labels=custom_labels, loc='upper left')
plt.savefig('../thesis/images/totalGold.pgf')
```
